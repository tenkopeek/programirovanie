```python
from math import comb
from decimal import getcontext, Decimal as dec

getcontext().prec = 50

# Метод Бернулли
def bern(n, k, p):
    return comb(n, k) * dec(p)**k * dec(1 - p)**(n - k)
```

## 3адание №1

```python
print(f"B = {bern(100000, 5800, 0.06):.6f}")
```

## 3адание №2

```python
n = 1200
p = 0.002

p0 = bern(n, 0, p)
p1 = bern(n, 1, p) 
p2 = bern(n, 2, p)
p3 = bern(n, 3, p)
print(f"¬B = {p0:.6f} + {p1:.6f} + {p2:.6f} + {p3:.6f} = {(p0 + p1 + p2 + p3):.6f}")
print(f"B = 1 - ¬B = 1 - {(p0 + p1 + p2 + p3):.6f} = {(1 - (p0 + p1 + p2 + p3)):.6f}")
```

## 3адание №3

```python
n = 1000
k = 14
p = 0.8

sum = 0
for i in range(k + 1):
    sum += bern(n, i, p)
print(f"B = {sum:.6e}")
```

## 3адание №4

```python
n = 30
p = 0.75

sum = 0
for i in range(25, n + 1):
    sum += bern(n, i, p)
print(f"¬B = {sum:.6f}")
print(f"B = 1 - ¬B = 1 - {sum:.6f} = {(1 - sum):.6f}")
```
